```python
#flask HTML link


app = Flask(__name__)



# Function to generate video (your existing model logic)
def generate_video(prompt):
    formatted_prompt = prompt + " in very anime like themed"

    torch.cuda.empty_cache()
    gc.collect()

    # Generate frames
    video_frames = pipe(formatted_prompt, num_inference_steps=50, height=320, width=576, num_frames=50).frames
    video_frames = video_frames.squeeze(0)

    # Export video
    video_path = "output.mp4"  # Temporary output file
    video_path = export_to_video(video_frames, fps=10)  # Make sure this function is defined

    # Ensure unique filename
    base_filename = prompt.replace(" ", "_")
    file_extension = ".mp4"
    counter = 1
    while os.path.exists(base_filename + str(counter) + file_extension):
        counter += 1
    unique_filename = base_filename + str(counter) + file_extension

    # Move the video file to the final destination
    final_path = os.path.join(os.getcwd(), unique_filename)
    shutil.move(video_path, final_path)

    return final_path





@app.route('/generate', methods=['POST'])
def generate():
    prompt = request.form.get('prompt')
    if not prompt:
        return jsonify({"error": "Prompt is required"}), 400

    video_path = generate_video(prompt)
    return send_file(video_path, as_attachment=True)  # Send video file to frontend

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
